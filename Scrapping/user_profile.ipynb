```python
from faker import Faker
import pandas as pd
import random
import phonenumbers

# Create a Faker instance
faker = Faker()

# Generate synthetic user data
num_users = 5000
user_data = []

for user_id in range(1, num_users + 1):
    # Generate a random country code
    country_code = random.choice(list(phonenumbers.COUNTRY_CODE_TO_REGION_CODE.keys()))

    # Generate a random phone number for the selected country
    fake_number = faker.numerify(text='##########')  # Generate a 10-digit fake number
    phone_number = phonenumbers.format_number(phonenumbers.parse(f"+{country_code}{fake_number}"), phonenumbers.PhoneNumberFormat.E164)

    user_data.append({
        'userid': user_id,
        'name': faker.name(),
        'age': random.randint(18, 60),
        'gender': random.choice(['Male', 'Female']),
        'email': faker.email(),
        'address': faker.address(),
        'phone_number': phone_number,
        'travel_purpose': random.choice(['Business', 'Leisure', 'Education', 'Other'])
        # Add other relevant fields as needed
    })

# Create a DataFrame
df = pd.DataFrame(user_data)

# Save the synthetic user profile data to a CSV file
df.to_csv('synthetic_user_profiles.csv', index=False)
```
