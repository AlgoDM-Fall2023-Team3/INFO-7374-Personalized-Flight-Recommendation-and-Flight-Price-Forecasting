```python
import pandas as pd
from pandas import json_normalize
import json

with open('/content/drive/MyDrive/ADM-FinalProject/data/flight_results_2.json') as f:
    flights_data = json.load(f)

flights_list = []
for f in flights_data:
    for result in f.get("results", []):
      content = result.get("content", {})
      content_res = content.get("results", {})
      flights = content_res.get("flights", {})
      flights_list.append(flights)

print(flights_list)
```

```python
flattened_data = json_normalize(flights_list, sep="_")

# Convert the flattened data to a DataFrame
df = pd.DataFrame(flattened_data)

# Display the DataFrame
df.head()
```

```python
new_df = df.explode('results')
new_df[['url','type','price','airline','duration','abc']] = ( new_df['results'].agg(pd.Series) )
new_df.drop(['results', 'abc'], axis =1, inplace = True)
new_df = new_df.dropna()
new_df
```

```python
combined_df = pd.concat([df,new_df])
combined_df.drop(['Unnamed: 0', 'abc'], axis = 1 , inplace = True)
combined_df = combined_df.dropna()
combined_df.head()
```

```python
combined_df.price.value_counts()
```

```python
combined_df['price'] = combined_df['price'].str.replace('$','')
combined_df['price'] = combined_df['price'].str.replace(',','')
combined_df['price'] = combined_df['price'].str.replace('unknown','0')
combined_df['price'] = combined_df['price'].astype(int)
combined_df['price']
```

```python
combined_df.dtypes
```

```python
from datetime import datetime

from dateutil import parser
import pandas as pd

current_year = datetime.now().year

combined_df['date_from'] = combined_df['date_from'].apply(lambda x: parser.parse(x + f', {current_year}'))
combined_df['date_to'] = combined_df['date_to'].apply(lambda x: parser.parse(x + f', {current_year}'))

combined_df['date_from'] = combined_df['date_from'].apply(lambda x: x.replace(year=2023) if x.month == 12 and x.day <= 31 else x.replace(year=current_year+1))
combined_df['date_to'] = combined_df['date_to'].apply(lambda x: x.replace(year=2023) if x.month == 12 and x.day <= 31 else x.replace(year=current_year+1))
```

```python
combined_df.head()
```

```python
combined_df['departure_day'] = combined_df['date_to'].dt.strftime('%A')  # Format as the full name of the day
#df['date_to_date'] = df['date_to'].dt.date

combined_df['arrival_day'] = combined_df['date_from'].dt.strftime('%A')  # Format as the full name of the day
#df['date_from_date'] = df['date_from'].dt.date
```
